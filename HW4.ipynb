# Уровень 0
# посчитайте значение 1-ой и 2-ой производной фунции
# f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pylab

from scipy.misc import derivative
```

```python
from scipy.misc import derivative
```

```python
def f(x):
    return x**5 + 4*np.sin(2*x)+np.cos(3*x+3)
derivative(f, 1.0, dx=1e-6)
```

```python
derivative(f, 1.0, n=2, dx=1e-6)
```

# Уровень 1**
# Постройте график первой и второй производной функции  
(sin(2x+1))^5  
на отрезке [-5; 5]

```python
def f_2(x):
    return (np.sin(2*x+1))**5
```

```python
x1 = np.linspace(-5, 5, 50)
y1 = [derivative(f_2, i, dx=1e-6) for i in range(len(x1))]
plt.plot(x1, y1, '-o')
plt.title('График первой производной функции')
plt.ylabel("y'")
plt.xlabel('x')

plt.show()
```

```python
x1 = np.linspace(-5, 5, 50)
y1 = [derivative(f_2, i, n=2, dx=1e-6) for i in range(len(x1))]
plt.plot(x1, y1, '-o')
plt.title('График второй производной функции')
plt.ylabel("y''")
plt.xlabel('x')

plt.show()
```
